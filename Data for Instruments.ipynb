## Guitar

```python
f_g = au.read_wav_mono('audio_files/guitar/guitar_A4_very-long_forte_normal.wav')
au.graph_signal(f_g)
print(ap.fundamental(f_g))
```

```python
step= int(44100 / 4)
start = 11025
chunk_g = f_g[start:start+step]
s_g = fft.rfft(chunk_g)
au.graph_fft(s_g, 6000)
```

```python
w_g = ap.hann_window(chunk_g)
s_g = fft.rfft(w_g)
au.graph_fft(s_g, 10000)
print(ap.spectral_centroid(s_g))
```

```python
s_g = ap.spectral_flux(f_g)
au.graph_signal(s_g, 16)
```
