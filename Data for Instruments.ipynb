```python
import audio_processing as ap
import audio_utils as au
import math
import numpy.fft as fft
import matplotlib.pyplot as plt
%matplotlib inline
```

## Guitar

```python
f_g = au.read_wav_mono('audio_files/guitar/guitar_A4_very-long_forte_normal.wav')

au.graph_signal(f_g)
print(ap.fundamental(f_g))
```

```python
flux_g = ap.spectral_flux(f_g)
au.graph_signal(flux_g, 16)
```

```python
step= int(44100 / 4)
start = 11025
chunk_g = f_g[start:start+step]
s_g = fft.rfft(chunk_g)
au.graph_fft(s_g, 6000)
```

```python
w_g = ap.hann_window(chunk_g)
s_g = fft.rfft(w_g)
au.graph_fft(s_g, 6000)
```

```python
s_g = ap.spectral_flux(f_g)
au.graph_signal(s_g, 100)
```
