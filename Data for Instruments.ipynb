```python
import audio_processing as ap
import audio_utils as au
import math
import numpy.fft as fft
import matplotlib.pyplot as plt
%matplotlib inline
```

## Guitar

```python
f_g = au.read_wav_mono('audio_files/guitar/guitar_A4_very-long_forte_normal.wav')
au.graph_signal(f_g)
print(ap.fundamental(f_g))
```

```python
step= int(44100 / 4)
start = 44100
chunk = f_g[start:start+step]
s = fft.rfft(chunk)
au.graph_fft(s, 1000)
```

```python
w = ap.hann_window(chunk)
s = fft.rfft(w)
au.graph_fft(s, 2000)
```

```python
s = ap.spectral_flux(f_g)
au.graph_signal(s)
```
